```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option("display.expand_frame_repr", False)
```

```python
#open excel file demographics
df = pd.read_excel(folder+'demographics-nhs-indicators-2011.xlsx').copy()
df.head()
```

```python
#use first row as header
header = df.columns
df.columns=header.str.lstrip(' ')
df = df.iloc[1:,:]
df.head()
```

```python
#create new columns that indicate percentage of features
df['minority_perc'] = df['Total visible minority population']/df['Total Population']
df['South Asian%'] = df['South Asian']/df['Total Population']
df['Chinese%'] = df['Chinese']/df['Total Population']
df['Black%'] = df['Black']/df['Total Population']
df['west_asian%']= df['West Asian']/df['Total Population']
df['latin_american%'] = df['Latin American'] /df['Total Population']
df['unemployment%'] = df['Unemployed'] / df['Total Population']
df['no_college'] = df['No certificate, diploma or degree'] /df['Total Population']
df['owner%']= df['Owner'] / (df['Owner']+df['Renter'])
df['renter%'] = df['Renter']/(df['Owner']+df['Renter'])
```

```python
df.head()
```

```python
#drop extra not helpful columns
df_demo = df.drop(df.loc[:,'Child0-14':'Renter'], axis=1)
df_demo.drop(columns=['Major repairs needed','Median after-tax household income $'],inplace=True)
df_demo.head()
```

```python
#read new excel file and use first row as header
df_eco = pd.read_excel(folder+'wellbeing-toronto-economics.xlsx').copy()
header = df_eco.iloc[0,:]
header
df_eco.columns = header
df_eco = df_eco.iloc[1:,:]
df_eco
```

```python
#combine two dataframes
table = df_demo.merge(df_eco, how='inner', left_on='Hood#', right_on='Neighbourhood Id',copy=True, sort= True)
table.head()
table.drop(columns=['Neighbourhood Id'], inplace=True)
```

```python
table.to_csv(folder+'economic_hoods.csv')
```
